### K Means Cluster Creation

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=2)
```

```python
kmeans.fit(df.drop('Private',axis=1))
```

```python
kmeans.cluster_centers_
```

### Evaluation

```python
# creating a new colum Cluster, which is a 1 for a Private school, and a 0 for a public school
def cluster_convert(cluster):
    if cluster == 'Yes':
        return 1
    else:
        return 0
```

```python
df['Cluster'] = df['Private'].apply(cluster_convert)
```

```python
df.tail(2)
```

```python
df.info()
```

```python
# to see how well the Kmeans clustering worked without being given any labels
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(confusion_matrix(df['Cluster'],kmeans.labels_))
```

```python
print(classification_report(df['Cluster'],kmeans.labels_))
```
